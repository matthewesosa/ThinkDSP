```python
import numpy as np
from numpy import poly, array
import matplotlib.pyplot as plt
from thkdsp.ztransform import residue, residuez, zplane
from scipy.signal import lfilter
```

```python
n = lrange(-5, 20)
```

```python
sn = np.sin(2*np.pi*n/4) * step(n)
```

```python
zn1 = np.complex(0, 1)
```

```python
zn2 = np.complex(0, -1)
```

```python
zp1 = np.complex(-1/2, 1/2)
```

```python
zp2 = np.complex(-1/2, -1/2)
```

```python
a = 5/4 * poly([zn1, zn2])
```

```python
b = poly([zp1, zp2])
```

```python
fig, ax = zplane(a, b) #pole-zero plot
```

```python
r, p, k = residue(a, b)
```

```python
print('r = '+str(r)+', p = '+str(p)+', k = '+str(k))
```

```python
gn = lfilter(a, b, sn)
```

```python
fig1, ax1 = plot_signal(n, sn, xlabel = "n", ylabel = "s[n]", color = "r") #input signal
```

```python
fig2, ax2 = plot_signal(n, gn, xlabel = "n", ylabel = "g[n]", color = "r") #output signal
```
