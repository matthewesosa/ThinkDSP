### Using the matrix-vector multiplication approach discussed in the tutorial, write a Python function to compute the DTFT of a finite-duration sequence. 

```python
def dtft(n, s, F):
    M = len(F)
    N = len(s)
    if (len(n) != N):
        print("use same length for n and s")
        return None
    
    S = s.dot(exp(-2j*pi * n.reshape(N, 1).dot(F.reshape(1, M))))
    return S   
    
```

```python
# Using a frequency location vector F of 1000, equidistant sampling points between –1/2 and 1/2

F = lrange(-0.5, 0.5, 1000)
```

### a) s[n] = (0.5)^|n| [ε[n+10]−ε[n−11]].
    

```python
n = lrange(-10,11)
s = (0.5)**abs(n) * (step(n+10) - step(n-11))
```

```python
S = dtft(n, s, F)
```

```python
mag = abs(S)
ang = angle(S)
```

```python
plot_signal(n, s, 'n','s')
plot_signal(F, mag, xlabel = 'F',  ylabel = 'Sa')
plot_signal(F, ang, 'F', 'ϕ')
```

Comment on the angle plot:

The angle plot above agrees with the fact that the phase of the DTFT of real valued sequences is always odd.
