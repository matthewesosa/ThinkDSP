```python
from thkdsp import *
from numpy import *
from thkdsp.dtft import dtft
```

```python
#For the normalized frequency variable F use a frequency location vector F of 1000 (or 1001) 
#equidistant sampling points between –1/2 and 1/2
F = lrange(-1/2, 1/2, 1001)
```

Solving Question a,

```python
#Question A
#s[n] = (0.5) |n| [ε[n + 10] − ε[n − 11]]

n1 = lrange(-10, 11)

s=(0.5**abs(n1))*(step(n1+10)-step(n1-11))
```

```python
S = dtft(n1, s, F)
```

```python
angle = np.angle(S)
```

Plotting

```python
plot_signal(n1, s, isdiscrete=True, xlabel='n', ylabel='s')
plot_signal(F,abs(S),xlabel='F',ylabel='Sa_F')
plots.plot_signal(F,angle,xlabel='F',ylabel='Angle')
```

```python
#Comment on the angle plot: The angle displayed sets wof waveforms changing rapidly, 
#and it appears very small near zero.
```
