```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# My Data
# using three log with distance 100m


# Defining data set for throughput and time
runTime = np.array([15,      25,     35,      45,      55,      65,      75,   85])
Throughput = np.array([68.7701, 71.2401, 72.262, 72.8055, 73.1498, 73.3993, 73.5584, 73.6824 ])


#Throughput = np.array([67.9486, 72.422, 73.8099, 74.4867, 74.8877, 75.1517, 75.3401, 75.4799]) 

plt.plot(runTime, Throughput)
plt.scatter(runTime, Throughput)

plt.grid(True)                 
plt.xlabel('Simulation run time in seconds')
plt.ylabel('Throughput Mbps')
plt.savefig('Sim_time.png', dpi=500, bbox_inches='tight')

plt.tight_layout()
plt.show()



#We Concluded on a Simulation time of 80secs
```

```python
# through put vs dist

dista = np.array([1,       50,      100,       200,     500,  1000,     2000,     4000,   6000,    9000])
friis = np.array([75.4799, 75.4796, 48.0041, 28.217, 0, 0, 0, 0, 0, 0])
fixed = np.array([36.3387, 36.3361, 36.3334, 36.3307, 36.3172, 36.2929, 36.2503, 35.0673, 35.0053, 0])

three = np.array([75.4799, 75.4802, 73.6257, 36.2903, 0, 0, 0, 0, 0, 0])
two   = np.array([75.4799, 75.4796, 35.5003, 0, 0, 0, 0, 0, 0, 0,])
naka   = np.array([75.4795, 75.4792, 75.4793, 75.4793, 75.479, 75.4793, 75.4792, 75.4795, 75.4793, 0.000299342])







fig, ax = plt.subplots()

ax.plot(dista, friis, label= 'FriisPropagationLossModel', color= 'orange')
ax.plot(dista, fixed, label= 'FixedRssLossModel', color= 'red')
ax.plot(dista, three, label= 'ThreeLogDistancePropagationLossModel', color= 'blue')
ax.plot(dista, two,   label= 'TwoRayGroundPropagationLossModel', color= 'green')
ax.plot(dista, naka,  label= 'NakagamiPropagationLossModel', color= 'black', linestyle='--')  

ax.legend()
ax.yaxis.grid(True)

#plt.title('\nA plot showing the varying UDP throughput for the five different propagation models \nover varying distance\n')
ax.xaxis.set_label_text('Distance in meters')
ax.yaxis.set_label_text('Throughput (Mbps)')
         
plt.xlim([0, 800])
plt.ylim([0, 80])
plt.savefig('thput.png', dpi=500, bbox_inches='tight')
plt.tight_layout()

plt.show()
```
