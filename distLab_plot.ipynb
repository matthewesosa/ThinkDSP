```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Defining data set for throughput and time
runTime = np.array([10,      20,     30,      40,      50,      60,      70,      80])
Throughput = np.array([67.9486, 72.422, 73.8099, 74.4867, 74.8877, 75.1517, 75.3401, 75.4799]) 

plt.plot(runTime, Throughput)
plt.scatter(runTime, Throughput)

plt.grid(True)                 
plt.xlabel('Simulation run time in seconds')
plt.ylabel('Throughput Mbps')
plt.savefig('time.png', dpi=500, bbox_inches='tight')

plt.tight_layout()
plt.show()
```

```python
# Defining data set for propagation model and distance.
dista = np.array([1,       10,      50,       100,     150,     225,     250,     300,     500,     1000,   5000,    10000])
friis = np.array([75.0225, 75.0225, 75.0223,  47.3038, 35.9772, 23.8636, 0,       0,       0,       0,      0,       0])
fixed = np.array([75.0225, 75.0225, 75.0225,  75.0221, 75.0223, 75.0227, 75.0223, 75.0223, 75.0221, 75.0223,75.0225, 0])
three = np.array([75.0225, 75.0225, 75.0225,  18.2611, 48.1141, 2.67904, 0,       0,       0,       0,      0,       0])
two   = np.array([75.0225, 75.0225, 75.0225,  47.3038, 35.9772, 0,       0,       0,       0,       0,      0,       0])
naka  = np.array([75.0211, 75.0211, 75.0225,  75.0215, 75.0211, 75.0215, 75.0215, 75.0213, 75.0215, 75.0215,75.0213, 0.000801627])

fig, ax = plt.subplots()

ax.plot(dista, friis, label= 'FriisPropagationLossModel', color= 'orange')
ax.plot(dista, fixed, label= 'FixedRssLossModel', color= 'red')
ax.plot(dista, three, label= 'ThreeLogDistancePropagationLossModel', color= 'gray')
ax.plot(dista, two,   label= 'TwoRayGroundPropagationLossModel', color= 'green')
ax.plot(dista, naka,  label= 'NakagamiPropagationLossModel', color= 'black', linestyle='--')  

ax.legend()
ax.yaxis.grid(True)

#plt.title('\nA plot showing the varying UDP throughput for the five different propagation models \nover varying distance\n')
ax.xaxis.set_label_text('Distance in meters')
ax.yaxis.set_label_text('Throughput (Mbps)')
         
plt.xlim([0, 800])
plt.ylim([0, 80])
plt.savefig('tp.png', dpi=500, bbox_inches='tight')
plt.tight_layout()

plt.show()
```
