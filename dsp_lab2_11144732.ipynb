### Moving average of white noise :

```python
# Signal duration T_d and Sampling rate r

T_d = 0.001
r = 48000
```

```python
t_d, d = gen_dirac(T_d, r)
```

```python
# Impulse respone


t_h, h = moving_average(t_d, d, 1/r)
```

```python
# Plot of the impulse response

plot_signal(t_h, h, ylabel="h")
```

```python
# Generating a sample function of the output process
t_g, g = moving_average(t_s, s, 1.0/4800.0)
```

```python
# Plot of the white noise input process and the filtered noise output process 

signal_fig, signal_ax = plot_signal(t_s, s, xmin=0, xmax=0.05, plotaxes=False)
plot_signal(t_g, g, xmax=0.05, figure=signal_fig, axes=signal_ax, color="r")
leg = signal_ax.legend(('s(t)', 'g(t)'))
```

##### There is one obvious difference between the input and the output process :   

The ouput process has a reduced amplitude (reduced average power); it is not changing as fast as the input process because the Moving Average operation acts like a simple low pass filter that reduces the contribution of higher frequencies.


```python
# Zoom in a bit moreto spot a second less obvious difference:


signal_fig, signal_ax = plot_signal(t_s, s, xmin=0.01, xmax=0.015, plotaxes=False)
plot_signal(t_g, g, xmin=0.01, xmax=0.015, figure=signal_fig, axes=signal_ax,color="r")
```

###### Comment on the second difference between the input and the output process:

Some higher frequencies are still present and visible as little spikes in the output process.
